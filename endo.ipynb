```python
df = df.drop(columns=['Unnamed: 32'])  # Remove useless column

y = df['diagnosis']  # Correct target variable
x = df.drop(columns=['diagnosis'])  # Features for training
```

```python
print("Shape of x before split:", x.shape)
print("Shape of y before split:", y.shape)
```

##data separation

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)

print(x_train.shape, x_test.shape)  # Should be (455, 31) and (114, 31)
print(y_train.shape, y_test.shape)  # Should be (455,) and (114,)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1️⃣ Scale the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# 2️⃣ Train the model with more iterations
clf = LogisticRegression(max_iter=500)  # Increase max_iter
clf.fit(x_train_scaled, y_train)

# 3️⃣ Make predictions
y_pred = clf.predict(x_test_scaled)

# 4️⃣ Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
